```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(0)  # For reproducibility
df = pd.DataFrame({
    'Accident_ID': range(1, 1001),
    'Road_Condition': np.random.choice(['Dry', 'Wet', 'Ice', 'Snow'], 1000),
    'Weather': np.random.choice(['Sunny', 'Cloudy', 'Rainy', 'Foggy'], 1000),
    'Time_of_Day': np.random.choice(['Morning', 'Afternoon', 'Evening', 'Night'], 1000),
    'Location': np.random.choice(['Urban', 'Rural', 'Highway'], 1000),
    'Latitude': np.random.uniform(37.0, 42.0, 1000),
    'Longitude': np.random.uniform(-122.0, -117.0, 1000),
    'Number_of_Vehicles': np.random.randint(1, 6, 1000),
    'Number_of_Injuries': np.random.randint(0, 10, 1000),
    'Number_of_Fatalities': np.random.randint(0, 5, 1000)
})
```

```python
df
```

```python
# Count of accidents by road condition
road_condition_counts = df['Road_Condition'].value_counts()
print("Accidents by Road Condition:")
print(road_condition_counts)

# Plot accidents by road condition
plt.figure(figsize=(8, 6))
sns.countplot(x='Road_Condition', data=df)
plt.title("Accidents by Road Condition")
plt.show()
```

```python
# Count of accidents by weather
weather_counts = df['Weather'].value_counts()
print("\nAccidents by Weather:")
print(weather_counts)

# Plot accidents by weather
plt.figure(figsize=(8, 6))
sns.countplot(x='Weather', data=df)
plt.title("Accidents by Weather")
plt.show()
```

```python
# Count of accidents by time of day
time_of_day_counts = df['Time_of_Day'].value_counts()
print("\nAccidents by Time of Day:")
print(time_of_day_counts)

# Plot accidents by time of day
plt.figure(figsize=(8, 6))
sns.countplot(x='Time_of_Day', data=df)
plt.title("Accidents by Time of Day")
plt.show()
```

```python
# Count of accidents by location
location_counts = df['Location'].value_counts()
print("\nAccidents by Location:")
print(location_counts)

# Plot accidents by location
plt.figure(figsize=(8, 6))
sns.countplot(x='Location', data=df)
plt.title("Accidents by Location")
plt.show()
```

```python
# Plot accident hotspots
plt.figure(figsize=(8, 6))
plt.scatter(df['Longitude'], df['Latitude'], s=5, alpha=0.5)
plt.title("Accident Hotspots")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
# Correlation matrix
corr_matrix = df[['Number_of_Vehicles', 'Number_of_Injuries', 'Number_of_Fatalities']].corr()
print("\nCorrelation Matrix:")
print(corr_matrix)

# Plot correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
plt.title("Correlation Matrix")
plt.show()
```
